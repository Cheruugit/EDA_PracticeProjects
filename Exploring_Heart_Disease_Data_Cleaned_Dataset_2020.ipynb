# Unveiling Heart Health: Discovering Patterns and Trends Through Data Exploration

Let's dive into our Heart Disease data to learn more about it! Exploratory Data Analysis (EDA) is like a detective mission where we uncover clues about each piece of information. ## About dataset

HeartDisease (Target):A binary trait indicating the presence or absence of heart disease.

BMI (Body Mass Index):
A numerical value assessing the relationship between a person's mass and height, providing insight into weight status.
Smoking:A binary variable indicating whether an individual is a smoker, a significant risk factor for cardiovascular disease. Smoking affects heart rate, blood pressure, and blood clotting.

AlcoholDrinking:A binary variable indicating alcohol consumption, which can lead to both temporary and permanent heart-related issues.

Stroke:A binary trait indicating the occurrence of stroke, with a focus on ischemic stroke caused by heart-related factors such as impaired heart function and thromboembolism.

PhysicalHealth:A count of the number of days in a month an individual experienced poor physical health.

MentalHealth:A count of the number of days in a month an individual experienced poor mental health.

DiffWalking:A binary variable indicating difficulty climbing stairs, reflecting mobility issues.

Sex:Categorical variable indicating the gender of the individual.

AgeCategory:Categorical variable representing age categories of the subjects.

Race:Categorical variable indicating the race of the individual.

Diabetic:Binary variable indicating whether an individual has diabetes.

PhysicalActivity:Binary variable indicating whether an individual engaged in physical activity or exercise in the past 30 days.

GenHealth:Categorical variable representing the general well-being of the individual.

SleepTime:Numerical variable indicating the number of hours of sleep an individual gets.

Asthma:Binary variable indicating whether an individual has asthma.

KidneyDisease:Binary variable indicating whether an individual has kidney disease.

Skin Cancer:Binary variable indicating whether an individual has skin cancer.

  Since AgeCategory is a categorical variable, we will have to use a countplot.

```python
# Define the order of age categories
age_order = df['AgeCategory'].value_counts().index.sort_values()
sns.countplot(data=df, x='AgeCategory', palette='viridis', order=age_order)
plt.title('Distribution of Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

To use a histogram effectively with age categories, we`ll create a new column representing the midpoint of each category. By calculating the average of each category, the data becomes numerical and continuous, making it easier for the histogram to interpret.

```python
def calculate_midpoint(age_category):
    lower_bound, upper_bound = map(int, age_category.split('-'))
    return (lower_bound + upper_bound) / 2
df['Midpoint'] = df['AgeCategory'].apply(calculate_midpoint)

```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Midpoint'], color='lightgreen', edgecolor='black')
plt.title('Distribution of Age Midpoint')
plt.xlabel('AgeGroup')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

## Bivariate Analysis

```python
#Does the difficulty of walking correlate with age
grouped_df = df.groupby(['AgeCategory', 'DiffWalking']).size().unstack()
# Create a stacked bar plot of heart disease incidence by age category
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', stacked=True, color=["lightgreen", "darkviolet"]
)
plt.title('Relationship between difficulty of walking with Age')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.legend(title='DiffWalking', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the chart we can observe that as age advances, there appears to be a corresponding increase in the frequency of reported difficulty with walking suggesting a potential association between age and mobility challenges.

```python
#Does the difficulty of walking correlate with BMI?
sns.violinplot(x='DiffWalking', y='BMI', data=df)
plt.title('Relationship between BMI and Difficulty of Walking')
plt.show()
```

It can be observed that individuals without difficulty walking tend to have a lower average BMI compared to those with difficulty walking, we can compare the average BMI values for each group to support this observation.

```python
# Calculate average BMI for each group
average_bmi_diffwalking_yes = df[df['DiffWalking'] == 'Yes']['BMI'].mean()
average_bmi_diffwalking_no = df[df['DiffWalking'] == 'No']['BMI'].mean()
print('Average BMI')
print(f' with difficulty walking (DiffWalking = Yes): {average_bmi_diffwalking_yes:.2f}')
print(f' without difficulty walking (DiffWalking = No): {average_bmi_diffwalking_no:.2f}')
```

 This confirms the observation that individuals with difficulty walking ('Yes') have a higher average BMI (31.20) compared to those without difficulty walking ('No') with a lower average BMI (27.86).

```python
#Is there a relationship between alcohol drinking and the occurrence of strokes?
sns.countplot(x='AlcoholDrinking', hue='Stroke', data=df, palette= 'pastel')
plt.title('alcohol drinking vs the occurrence of strokes')
```

While the initial analysis suggests a higher occurrence of strokes among those who do not drink alcohol, it's important to acknowledge the imbalance in the dataset, with 93.2% of the population abstaining from alcohol.

```python
ct_percentage = pd.crosstab(df['AlcoholDrinking'], df['Stroke'], normalize='index') * 100

# Plot a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(ct_percentage, annot=True, fmt='.2f', cmap='Blues', cbar=True, linewidths=.5)
plt.title('Percentage of Stroke Occurrence based on Alcohol Drinking')
plt.xlabel('Stroke')
plt.ylabel('Alcohol Drinking')
plt.show()
```

From the dataset, it appears that 2.37% of individuals who drink alcohol are likely to experience a stroke, while a slightly higher percentage, 3.88%, is observed among those who do not consume alcohol. These percentages highlight the importance of considering both the prevalence of stroke within each group

## Conclusions and Recommendations

Our analysis underscores the profound impact of heart disease on both physical and mental health outcomes. Individuals with heart disease endure prolonged periods of poor physical and mental health compared to their healthier counterparts. This highlights the urgent need for interventions targeting these areas to enhance overall well-being among heart disease patients.

Moreover, our findings reveal a positive correlation between physical activity levels and physical health. As individuals engage in higher levels of physical activity, we observe tangible improvements in their physical well-being. This underscores the importance of promoting regular physical activity as a pivotal component of heart disease management and prevention strategies.

Increased days of poor mental health appear to correlate with a slight reduction in sleep time. This suggests a potential link between mental well-being and sleep patterns, highlighting the importance of addressing both areas for overall health improvement. However further research is needed to explore the causal relationship between poor mental health and decreased sleep time fully.

Given these insights, we advocate for the implementation of targeted interventions aimed at improving physical and mental health outcomes among individuals with heart disease. These interventions should encompass structured exercise programs, stress management techniques, and accessible mental health support services. Additionally, fostering a culture of regular physical activity within communities can serve as a proactive measure against heart disease onset, contributing to overall health and vitality. By prioritizing these areas, we can effectively enhance the quality of life for heart disease patients and cultivate a healthier society at large.
