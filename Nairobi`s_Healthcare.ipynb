# **Enhancing Healthcare Accessibility in Nairobi**

Improving access to healthcare services is essential for better health outcomes and overall well-being, especially in growing urban areas like Nairobi County.
However, the strain on the already existing healthcare infrastructure means that essential medical care becomes inaccessible to many. Despite being vital hubs for community health, healthcare facilities in Nairobi County face challenges in ensuring equitable access for all residents.

In this case study, we explore strategies to enhance healthcare accessibility in Nairobi, aiming to ensure that every resident can access the care they need.


The aim of this analysis is to identify gaps and opportunities in healthcare services offered across different areas of Nairobi, and to identify areas with limited access to healthcare.

## **ABOUT DATA**

Facility Code: A unique identifier for each health facility.

Facility Name: The name or title of the health facility.

Province: The administrative division within a country where the facility is located.

County: Another administrative division within a country, typically smaller than a province.

District: A further subdivision within a county or province, often used for administrative purposes.

Division: Potentially another level of administrative subdivision, depending on the country's administrative structure.

Type: The classification or category of the health facility (e.g., hospital, clinic, health center).

Owner: The entity or organization that owns or operates the health facility (e.g., government, private, NGO).

Location: The geographical coordinates or address of the facility's physical location.
Sub Location: A more detailed geographical description, potentially indicating a specific area within a location.

Description of Location: Additional details or descriptors about the facility's location.

Constituency: A geographical electoral area, often used in political contexts.

Nearest Town: The nearest town or urban center to the facility.

Beds: The number of hospital beds available in the facility.

Cots: The number of cots available, which might indicate capacity for temporary or short-term care.

Official Landline: The official telephone number for the facility.

Official Fax: The official fax number for the facility.

Official Mobile: The official mobile or cell phone number for the facility.

Official Email: The official email address for the facility.

Official Address: The official postal address of the facility.

Official Alternate No: An alternative contact number for the facility.

Town: The town or urban area where the facility is located.

Post Code: The postal code or ZIP code for the facility's address.

In Charge: The person or position responsible for the facility.

Job Title of in Charge: The job title or role of the person in charge.

Open 24 Hours: Indicates whether the facility operates 24 hours a day.

Open Weekends: Indicates whether the facility operates on weekends.

Operational Status: The current operational status of the facility (e.g., open, closed, under construction).

Service Availability: Various services available at the facility, including ANC (Antenatal Care), ART (Antiretroviral Therapy), OPD (Outpatient Department)


C-IMCI (Community Integrated Management of Childhood Illnesses), focusing on the integrated management of childhood illnesses at the community level.

Family Planning (FP) services, supporting reproductive health and family planning needs.

Home-Based Care (HBC), indicating support for patients who require care in their homes, often for chronic illnesses or palliative care.

Inpatient Department (IPD) services, which typically include admission and treatment for patients requiring overnight stays or specialized care.

## **DATA PREPROCESSING**

**Check for missing values**

```python
nairobi_facilities.isnull().sum()
```

Some columns in the dataset have no values present, indicating that certain essential healthcare services are missing within Nairobi, specifically on ( 'ANC', 'BEOC', 'BLOOD', 'CAES SEC','HCT', 'OPD', 'OUTREACH', etc).The absence on these services poses challenges on mother and child wellbeing, disease detection and management and blood shortages. This shows a major need of intervention within Nairobi to ensure the wellbeing of individuals at all stages of their lives.

```python
# Define the columns to drop
columns_to_drop = ['ANC', 'BEOC', 'BLOOD', 'CAES SEC', 'CEOC', 'EPI',
                   'GROWM', 'HCT', 'OPD', 'OUTREACH', 'PMTCT', 'RAD/XRAY',
                   'RHTC/RHDC', 'TB DIAG', 'TB LABS', 'TB TREAT', 'YOUTH']

nairobi_facilities = nairobi_facilities.drop(columns=columns_to_drop)
```

```python
nairobi_facilities.columns
```

```python
nairobi_facilities.isnull().sum()
```

**Services offered within Nairobi**

```python
sel_columns=['Open 24 Hours', 'Open Weekends' ,'ART', 'C-IMCI', 'FP', 'HBC', 'IPD']
# Check unique values in each column
for column in sel_columns:
    unique_values = nairobi_facilities[column].unique()
    print(f"Unique values in {column}: {unique_values}")
```

```python
nairobi_facilities[sel_columns] = nairobi_facilities[sel_columns].fillna('N')

# Check unique values in each column
for column in sel_columns:
    unique_values = nairobi_facilities[column].unique()
    print(f"Unique values in {column}: {unique_values}")
```

```python
open_24_hours_counts = nairobi_facilities['Open 24 Hours'].value_counts()

# Count facilities offering the "Open Weekends" service in each district
open_weekends_counts = nairobi_facilities['Open Weekends'].value_counts()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

axes[0].pie(open_24_hours_counts, labels=open_24_hours_counts.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Distribution of Facilities Offering "Open 24 Hours" Service')

axes[1].pie(open_weekends_counts, labels=open_weekends_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Distribution of Facilities Offering "Open Weekends" Service')

# Adjust layout
plt.tight_layout()
plt.show()
```

21% of the facilities within Nairobi are open 24 hours and 54% remain open on weekends. This indicates a half of healthcare centers extend their operational hours beyond standard working hours/days, thus improving accessibility for individuals with busy schedules or urgent healthcare needs.

**Lets see the ratio of population to facilities.**

```python
subcounty_population_data = {
    'Subcounty': ['Dagoretti', 'Embakasi', 'Kamukunji', 'Kasarani', 'Kibra', 'Langata', 'Makadara', 'Mathare', 'Njiru', 'Starehe', 'Westlands'],
    'Population': [217651, 492476, 136670, 381234, 94199, 96698, 96369, 106522, 307642, 109173, 153818]
}

district_facilities_data = {
    'District': ['Dagoretti North', 'Dagoretti South', 'Embakasi Central', 'Embakasi East', 'Embakasi North', 'Embakasi South', 'Embakasi West', 'Kamukunji', 'Kasarani', 'Kibra', 'Langata', 'Makadara', 'Mathare', 'Roysambu', 'Ruaraka', 'Starehe', 'Westlands'],
    'Number of Facilities': [76, 29, 44, 27, 26, 27, 40, 59, 58, 93, 61, 55, 16, 83, 40, 136, 72]
}

subcounty_population_df = pd.DataFrame(subcounty_population_data)
district_facilities_df = pd.DataFrame(district_facilities_data)

subcounty_population_df['District'] = subcounty_population_df['Subcounty'].apply(lambda x: x.split()[0])

combined_district_facilities = district_facilities_df.groupby('District').sum().reset_index()

merged_data = pd.merge(subcounty_population_df, combined_district_facilities, on='District')

merged_data['Population_to_Facility_Ratio'] = merged_data['Population'] / merged_data['Number of Facilities']

merged_data
```

The high population-to-facility ratio within Mathare suggests that there may be insufficient healthcare facilities to adequately serve the population.

**Recommendations**

Prioritize the establishment and provision of missing essential healthcare services, such as maternity care, blood tests, and disease screenings, across Nairobi. This could involve government initiatives, partnerships with NGOs, and incentivizing private healthcare providers to offer these services. This will contribute to reducing maternal and child mortality rates.

Allocate resources to increase the number of healthcare facilities in Mathare to better serve its population. This could include building new clinics or expanding existing ones to meet the growing demand for healthcare services.

Encourage collaboration between public, private, and non-profit sectors to ensure comprehensive healthcare coverage in Nairobi. This could involve joint ventures, shared resources, and coordinated efforts to address healthcare challenges collectively.

Identify and address the reasons behind the non-operational status of healthcare facilities, particularly medical clinics and dispensaries. This may involve providing support, training, or incentives to encourage facility owners to reopen and operate their clinics effectively.

Prioritize Community Health Needs: Engage with local communities to understand their healthcare needs and preferences, and tailor healthcare services accordingly. This could involve community health assessments, outreach programs, and community-led initiatives to promote health and well-being.
